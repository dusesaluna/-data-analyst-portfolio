```python
import sqlite3
import pandas as pd
import os

# Database Setup
database_name = "gold_analysis.db"
conn = sqlite3.connect(database_name)
cursor = conn.cursor()


# Create Tables 
cursor.execute("""
    CREATE TABLE IF NOT EXISTS gold_prices (
                Date TEXT PRIMARY KEY,
                Open REAL,
                High REAL,
                Low REAL,
                Close REAL,
                Volume INTEGER
            )
        """)

cursor.execute("""
    CREATE TABLE IF NOT EXISTS dxy_prices (
            Date TEXT PRIMARY KEY,
            Open REAL,
            High REAL,
            Low REAL,
            Close REAL,
            Volume INTEGER,
            Change_Percent REAL
    )
""")

cursor.execute("""
    CREATE TABLE IF NOT EXISTS cpi_data (
        Date TEXT PRIMARY KEY,
        CPIAUCSL REAL
    )
""")

cursor.execute("""
    CREATE TABLE IF NOT EXISTS fed_funds_rate (
        Date TEXT PRIMARY KEY,
        FEDFUNDS REAL
    )
""")

cursor.execute("""
    CREATE TABLE IF NOT EXISTS treasury_10yr (
        Date TEXT PRIMARY KEY,
        Treasury_Yield REAL
    )
""")


conn.commit()

def load_csv_to_sqlite(csv_file, table_name, conn):
    df = pd.read_csv(csv_file)
    df.to_sql(table_name, conn, if_exists='replace', index=False)
    print(f"Data loaded into {table_name} successfully!")

load_csv_to_sqlite("gold_prices.csv", "gold_prices",conn)
load_csv_to_sqlite("US dollar Index Historical Data.csv", "dxy_prices", conn)
load_csv_to_sqlite("cpi.csv", "cpi_data", conn)
load_csv_to_sqlite("fed_funds_rate.csv", "fed_funds_rate", conn)
load_csv_to_sqlite("treasury_10yr_monthly_2020_2025.csv", "treasury_10yr",conn)


```

```python
def drop_first_two_rows(database_name, table_name):
    """Drops the first two rows from a table using rowid."""
    try:
        conn = sqlite3.connect(database_name)
        cursor = conn.cursor()

        # Find the minimum rowid
        cursor.execute(f"SELECT MIN(rowid) FROM {table_name}")
        min_rowid = cursor.fetchone()[0]

        # Delete the first two rows
        cursor.execute(f"DELETE FROM {table_name} WHERE rowid = {min_rowid}")
        cursor.execute(f"DELETE FROM {table_name} WHERE rowid = {min_rowid + 1}")

        conn.commit()
        print(f"First two rows deleted from {table_name}.")

    except sqlite3.Error as e:
        print(f"An error occurred: {e}")

    finally:
        if conn:
            conn.close()

# Example Usage:
database_name = "gold_analysis.db"
table_name = "gold_prices"
drop_first_two_rows(database_name, table_name)
```
