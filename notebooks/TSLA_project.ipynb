```python
open_series = df['Open']
Q1_open = open_series.quantile(0.25)
Q3_open = open_series.quantile(0.75)
IQR_open = Q3_open - Q1_open
lower_bound_open = Q1_open - 1.5 * IQR_open
upper_bound_open = Q3_open + 1.5 * IQR_open
outliers_open = open_series[(open_series < lower_bound_open) | (open_series > upper_bound_open)]

print("Outliers in 'Open' column (using IQR):\n", outliers_open)
```

```python
columns_to_analyze = ['Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume']

for col in columns_to_analyze:
    series = df[col]
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = series[(series < lower_bound) | (series > upper_bound)]

    print(f"Outliers in '{col}' column (using IQR):\n", outliers)
    print(f"Number of outliers in '{col}': {len(outliers)}\n")

    plt.figure(figsize=(8, 6))
    sns.boxplot(y=series)
    plt.title(f'Box Plot of {col} with Outliers')
    plt.ylabel(col)
    plt.show()

    plt.figure(figsize=(15, 6))
    plt.plot(df['Date'], series, label=col, color='skyblue')
    plt.scatter(df.loc[outliers.index]['Date'], outliers, color='red', label='Outliers')
    plt.axhline(upper_bound, color='r', linestyle='--', alpha=0.5, label=f'Upper Bound: {upper_bound:.2f}')
    plt.axhline(lower_bound, color='r', linestyle='--', alpha=0.5, label=f'Lower Bound: {lower_bound:.2f}')
    plt.title(f'{col} Over Time with IQR Outliers Highlighted')
    plt.xlabel('Date')
    plt.ylabel(col)
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
def identify_outliers_iqr(series):
    """Identifies outliers in a pandas Series using the IQR method."""
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = series[(series < lower_bound) | (series > upper_bound)]
    return outliers

volume_series = df['Volume']
volume_outliers = identify_outliers_iqr(volume_series)

plt.figure(figsize=(10, 6))
plt.hist(volume_series, bins=30, color='skyblue', alpha=0.7, label='Volume Distribution')
if not volume_outliers.empty:
    plt.vlines(volume_outliers, ymin=0, ymax=plt.ylim()[1], color='red', label='Outliers', linestyle='--')
plt.title('Histogram of Volume with Outliers')
plt.xlabel('Volume')
plt.ylabel('Frequency')
plt.legend()
plt.grid(axis='y', alpha=0.7)
plt.tight_layout()
plt.show()
```

Modle Building 

```python
#train-test
from sklearn.model_selection import train_test_split

features = ['SMA_10', 'EMA_10', 'RSI', 'MACD', 'MACD_signal',
            'Rolling_Mean_5', 'Rolling_Std_5', 'Close_lag1',
            'DayOfWeek', 'Month'] 

X = df[features]
y = df['Target']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)
```

```python
#train
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier()
model.fit(X_train, y_train)
```

```python
#predict
y_pred = model.predict(X_test)
```

Model Evaluation

```python
#***
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
import seaborn as sns
import matplotlib.pyplot as plt

# Evaluate model
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
print("\nClassification Report:\n", classification_report(y_test, y_pred))

# Confusion Matrix
cm = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(5, 4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Down', 'Up'], yticklabels=['Down', 'Up'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

```python
#Feature Importance
import matplotlib.pyplot as plt
import seaborn as sns

importances = model.feature_importances_
features_series = pd.Series(importances, index=features).sort_values(ascending=False)

plt.figure(figsize=(10,5))
sns.barplot(x=features_series, y=features_series.index)
plt.title("Feature Importance")
plt.show()
```

```python
#ROC curve
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt

y_prob = model.predict_proba(X_test)[:, 1]
fpr, tpr, thresholds = roc_curve(y_test, y_prob)
roc_auc = auc(fpr, tpr)

plt.figure()
plt.plot(fpr, tpr, label=f"ROC curve (area = {roc_auc:.2f})")
plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend(loc="lower right")
plt.title('Receiver Operating Characteristic')
plt.show()
```
